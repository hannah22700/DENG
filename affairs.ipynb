```python
import requests
import pandas as pd
import numpy as np

# Beispiel
# https://api.openparldata.ch/v1/affairs/?search=(Klima | Umwelt) & Schweiz&search_mode=boolean&limit=10 → Complex query with grouping

BASE_URL = "https://api.openparldata.ch/v1"

def fetch_affairs(search_term: str = "AI", limit: int = 100, offset: int = 0):
    response = requests.get(
        f"{BASE_URL}/affairs",
        params={
            "search": search_term,
            "search_mode": "partial",
            "sort_by": "-begin_date",
            "limit": limit,
            "offset": offset
        }
    )
    print(response.url)        # shows the exact URL being called
    # print(response.status_code)
    # print(response.json()) 
    response.raise_for_status()
    return response.json()

offset = 0
limit = 500
all_affairs = []

while True:
    data = fetch_affairs(search_term="Künstliche Intelligenz", limit=limit, offset=offset)
    items = data.get("data", [])
    all_affairs.extend(items)
    
    if not data["meta"]["has_more"]:  # ← correct key
        break
    offset += limit

# JSON → DataFrame
df = pd.json_normalize(all_affairs)


# Englische Types  
types = np.unique(df["type_harmonized.en"].dropna())

print(f"Total fetched: {len(all_affairs)}")
print(f"Types: {types}")

df["title.de"].dropna()

```

```python
# Fetch people of interests for affair
persons_of_interest = []

for affair in all_affairs:
    r = requests.get(affair["links"]["contributors"])
    contributors = r.json().get("data", [])
    persons_of_interest.extend(contributors)
```
