```python
import requests
import pandas as pd
from io import StringIO
```

```python
def get_Data(offset):
    url = "https://api.openparldata.ch/v1/persons?output_format=csv&offset=" + str(offset)
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.text
        return data
    else:

        return -1

df_combined = pd.DataFrame()

i = 0
while (True):
    data = get_Data(i)

    # API gibt resp code 200 zurück auch mit empty body... 
    if data == -1 or data.strip() == "":
        break
    else: 
        i = i + 500
        df = pd.read_csv(StringIO(data))
        df_combined = pd.concat([df_combined, df],ignore_index=True)
        df_combined
        print(i)


```
