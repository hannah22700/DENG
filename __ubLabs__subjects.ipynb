```python
business = spp.get_data('Business', Language='DE')
df_business = pd.DataFrame(business[:10])
df_business.head(20)
```

```python
dfs = []
for period in [51, 52]: # 2019-2023 / 2024 - 2027
    biz = spp.get_data('Business', Language='DE', SubmissionLegislativePeriod=period)
    dfs.append(pd.DataFrame(biz[:10]))
```

```python
df_business = pd.concat(dfs, ignore_index=True)
print(df_business.shape)
```

```python
df_business.head(10)
```

```python
def get_votes():
    votes = []
    for period in [51, 52]:
        data = spp.get_data('Vote', Language='DE', IdLegislativePeriod=period)
        votes.append(pd.DataFrame(data))
    
    return pd.concat(votes, ignore_index=True)

df_votes = get_votes()
```

```python
df_votes_tagged = df_votes.merge(
    df_business[['BusinessShortNumber', 'TagNames']],
    on='BusinessShortNumber',
    how='left'
)

df_votes_tagged[['BusinessShortNumber', 'BillTitle', 'TagNames']].head(10)
```

```python
print(f"Total votes: {len(df_votes_tagged)}")
print(f"With tags: {df_votes_tagged['TagNames'].notna().sum()}")
print(f"Without tags: {df_votes_tagged['TagNames'].isna().sum()}")
```

```python
# 1. Wir holen uns die Businesses (Motionen, Vorschläge, Initiativen, etc.)
dfs = []
for period in [51, 52]:
    biz = spp.get_data('Business', Language='DE', SubmissionLegislativePeriod=period)
    dfs.append(pd.DataFrame(biz))
df_business = pd.concat(dfs, ignore_index=True)

# 2. Votes
def get_votes():
    votes = []
    for period in [51, 52]:
        data = spp.get_data('Vote', Language='DE', IdLegislativePeriod=period)
        votes.append(pd.DataFrame(data))
    return pd.concat(votes, ignore_index=True)

df_votes = get_votes()

# 3. Wir mergen das
df_votes_tagged = df_votes.merge(
    df_business[['BusinessShortNumber', 'TagNames']],
    on='BusinessShortNumber',
    how='left'
)

df_votes_tagged[['BusinessShortNumber', 'BillTitle', 'TagNames']].head(10)
```
